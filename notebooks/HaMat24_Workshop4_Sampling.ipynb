# Random samples

*Task* 5.12 as an example of how we can take a random sample.

```python
#Step 1
#Import pandas and numpy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Step 2
#For exercise 5.12, we just need to generate values from 1-500
#Create a variable which contain the numbered 1-500 households, using a numpy array

data = range(0, 501)

#Bonus question, why do I include 501 in my variable?
```

The built in python range function generates values starting from the first parameter up to, but not including, the second parameter.

```python
#Step 3

# Transform the household variable to a pandas Series and call it population
population = pd.Series(data)

#Print with index
print(population.head(10))

# Print without index
print(population.head(10).to_string(index=False))
```

```python
#Step 4
#When we generate a sample, we get a new sample each time, but if we set a 'seed' we get the same sample.

# Set a random seed
np.random.seed(0)

# Generate a random sample of 100 households using pandas
sample = population.sample(n=25)

# Print the Series without index (and only the first 10 values)
print(sample.head(10).to_string(index=False))
```

```python
# Step 5
#I 5.12 genererer vi en uniform distribution, men mere often arbejder vi med en normal distribution
#Så lad os lige prøve at genere en population, der repræsenterer en normalfordeling
#her kan vi bruge numpy's np.random.normal() funktion

#Generate population dataset of 500
#Obs: loc = mean, scale = standard deviation

population = np.random.normal(loc=225, scale=10, size=500)

# Transform the population variable to a pandas Series
population = pd.Series(population)

#Herfra kan vi også generere en stikprøve

# Set a random seed
np.random.seed(0)

# Generate a random sample of 25 households using pandas
sample = population.sample(n=25)
```

```python
# Step 6
#Does our sample seem representative of our population?
#It can be good to visualize the sample when we have to answer that question

#Create histograms for population and sample
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].hist(population, bins=50)
ax[0].set_title("Population Histogram")
ax[0].set_xlabel("household")
ax[0].set_ylabel("Frequency")

ax[1].hist(sample, bins=10)
ax[1].set_title("Stikprøve Histogram")
ax[1].set_xlabel("household")
ax[1].set_ylabel("Frequency")

plt.show()
```

The number of cars in Aalborg is 100,000. The average price of the cars is DKK 150,000 incl
a standard deviation of 15,000. Based on this data, you must generate a sample of 100
cars in Python (see ex. 5.12 in Python).

```python
import numpy as np

# Given data
mean_price = 150000  # Average price of the cars in DKK
std_dev = 15000       # Standard deviation of the car prices in DKK
sample_size = 100     # Number of cars in the sample
population = 100000
# Generate a sample of car prices
car_prices_sample = np.random.normal(mean_price, std_dev, sample_size)
car_prices = np.random.normal(mean_price, std_dev, population)

# Display the generated sample
print(car_prices_sample)
```

```python
# Step 6
#Does our sample seem representative of our population?
#It can be good to visualize the sample when we have to answer that question

#Create histograms for population and sample
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].hist(car_prices, bins=50)
ax[0].set_title("Car price Histogram")
ax[0].set_xlabel("Cars")
ax[0].set_ylabel("Price")
ax[0].tick_params(axis='x', rotation=90)


ax[1].hist(car_prices_sample, bins=10)
ax[1].set_title("Car sample price Histogram")
ax[1].set_xlabel("Cars")
ax[1].set_ylabel("Price")
ax[1].tick_params(axis='x', rotation=90)


plt.show()
```
