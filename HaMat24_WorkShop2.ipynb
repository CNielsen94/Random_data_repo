###2.1 Give two examples each of nominal, ordinal, and interval data.

Here are the examples:

**Nominal Data (No natural order between categories):**
- Types of cuisine (e.g., Italian, Chinese, Indian, Mexican).
- Car brands (e.g., Toyota, Ford, BMW, Mercedes).

**Ordinal Data (Natural order between categories):**
- Education levels (e.g., High School, Bachelor's, Master's, PhD).
- Customer satisfaction ratings (e.g., Very Unsatisfied, Unsatisfied, Neutral, Satisfied, Very Satisfied).

**Interval Data (No true zero point):**
- Temperature in Celsius or Fahrenheit (e.g., 20°C, 35°C, 50°F, 70°F).
- IQ scores (e.g., 90, 100, 110, 120).

These examples illustrate the differences in the types of data: Nominal data categorizes without a natural order, ordinal data categorizes with a natural order but without a consistent scale between categories, and interval data has a consistent scale but no true zero point.


Here's an example of where it can be confusing (but methodically important) to distinquish between nominal and ordinal:

```python
import pandas as pd

# Creating a DataFrame with specified columns and rows
data = {'Numerical_representation': [1, 2, 3],
        'Family_member (nominal)': ['Mom', 'Dad', 'Child'],
        'Degree (ordinal)': ['Bachelor Degree', 'Master Degree', 'PhD']}
df = pd.DataFrame(data)

df
```

In this case, in order to use our data for statistical inference, we've had to create a numerical representation of our data points for Degree and Family. <br>
The conceptual line at which the numerical representation changes between nominal and ordinal, is whether or not we can determine a sort of 'natural' ranking between the different data points. This natural order exists in Degree, making it **ordinal**. However we can not point out a 'natural' ranking between data points in Family, making it **nominal**. <br>

### 2.2 Specify the Data Type for Each of the Following Data Examples

- **a. The number of kilometers run by joggers per week:**
  - *Data Type:* Interval Data
  - *Explanation:* This data is quantitative and measures distance run, which can be ordered on a scale with equal intervals. However, it has no true zero point (negative distance is not possible), which is characteristic of interval data.

- **b. Starting salary for Cand.merc graduates:**
  - *Data Type:* Interval Data (more specifically Ratio Data)
  - *Explanation:* Salaries are quantitative and have a true zero point (no salary). The differences and ratios between salary values are meaningful, making it fit as ratio data. However, within the context of the three specified types, it's closest to interval data with meaningful intervals.

- **c. The months in which the company's employees choose to take a holiday:**
  - *Data Type:* Nominal Data
  - *Explanation:* The months are simply names of categories and do not have a natural order or ranking. They represent distinct groups without quantitative significance, typical of nominal data.

- **d. Socio-economic status: high, medium, low:**
  - *Data Type:* Ordinal Data
  - *Explanation:* These categories represent a clear order (from high to low) but do not indicate specific quantitative differences or intervals between them, which is a defining characteristic of ordinal data.


### 2.3 Determine the Data Type for Each of the Following Data Examples

- **a. The weekly closing price of Amazon.com stock:**
  - *Data Type:* Interval Data
  - *Explanation:* The closing price of a stock is a numerical value on a continuous scale. The differences between values are meaningful, but the scale does not have a true zero point (a stock price can't be negative), making it interval data.

- **b. The month with the highest unemployment rate at Hotel D’Angleterre:**
  - *Data Type:* Nominal Data
  - *Explanation:* The months are simply names of categories without any inherent order or scale. They represent distinct groups and cannot be quantitatively compared, fitting the criteria for nominal data.

- **c. The size of the soda purchased by a sample of McDonald’s customers:**
  - *Data Type:* Ordinal Data (if categorized by size like Small, Medium, Large)
  - *Explanation:* If the soda sizes are categorized (e.g., Small, Medium, Large), they have a natural order but do not specify the exact volume, nor is the difference between sizes necessarily consistent. This ordered but non-quantitative nature makes it ordinal data.

- **d. The number of Range Rovers imported to Denmark in the last 10 years:**
  - *Data Type:* Interval Data (if treated as counts on a scale)
  - *Explanation:* The number of vehicles imported is a numerical count. While it can be quantitatively compared, it lacks a meaningful zero point within the context of this example (the count starts from zero but represents an amount, not an absence), making it best suited as interval data.

- **e. Legoland's annual revenue:**
  - *Data Type:* Interval Data
  - *Explanation:* Annual revenue is a numerical value that can be ordered and compared. However, since the scale does not have a true zero point (negative revenue is not an absence of revenue but indicates loss), it classifies as interval data.

If you look at the function: plt.subplot(1,2,1)

- The first number (1) indicates the number of rows of subplots in the figure. In this case, we want just one row (horizontal), since we have 2 visualizations we want to put next to each other
- The second number (2) indicates the number of columns (vertical) of subplots in the figure. Same reasoning as before, since we want 2 visualization next to each other, we need 1 row with 2 columns (one for each on the same row)
- The third number (1) specifies the position of the subplot being created in the grid layout defined by the first two numbers. So where in our 1x2 frame do we want it? In this case, we place it in position 1

```python
import numpy as np

# Assuming 'df3_56' is your DataFrame with columns 'Time' and 'Mark'
x = df3_56['Time']
y = df3_56['Mark']

# Creating the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y)

# Calculating the trend line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red')  # This adds the trend line

# Adding labels and title
plt.xlabel('Time to Complete Exam (minutes)')
plt.ylabel('Mark')
plt.title('Relationship Between Exam Completion Time and Student Performance')

# Show grid and plot
plt.grid(True)
plt.show()
```
